# Introduction

Vous êtes un groupe d'hackers malveillants qui a pu récupérer des données clients d'une compagnie d'assurance. Cette compagnie d'assurance a de très mauvaises pratiques de protection des données et les données que vous avez réussi à récupérer contiennent des informations extrêmement sensibles comme le numéro de carte de crédit du client.

Néanmoins, ces données ne contiennent pas le nom et prénom de la personne, informations indispensables pour utiliser leur carte de crédit pour faire des achats sur Internet, vous devez donc trouver cette information autre part.

Afin de remédier à ce problème, vous avez réussi à hacker les données du gouvernement regroupant tous les foyers fiscaux de France, ce qui vous permettrait d'identifier les clients de la compagnie d'assurance si vous êtes malins. Un hacker de votre équipe a réussi à filtrer les foyers fiscaux pour qu'ils ne contiennent que les foyers des clients de l'assurance, ce qui vous simplifiera la tâche plus tard.

# Données

Vous disposez de trois jeux de données : 
* `"clients_exo.csv"` : Les données client volées de la compagnie d'assurance.
* `"foyers_df.csv"`: Les données des foyers fiscaux filtrées pour ne contenir que les foyers des clients de l'assurance.
* `"foyers_zipcode.json"` : Un dictionnaire vous permettant de relier un foyer fiscal au département dans lequel le foyer réside.

```python
# add zip code to clients_exo from address
def get_zipcode(address):
    return address.split("\n")[1][:2]

clients_exo["zipcode"] = clients_exo["Address"].apply(get_zipcode)

clients_exo
```

```python
from datetime import datetime, timedelta

def calculate_birth_date(driving_license_date, driving_license_age):
    # Convert the driving license date to a datetime object
    license_date = datetime.strptime(driving_license_date, '%Y-%m-%d')

    # Calculate the birth date by subtracting the age from the license acquisition date
    birth_date = license_date - timedelta(days=365 * driving_license_age)

    # Format the birth date as YYYY-MM-DD
    formatted_birth_date = birth_date.strftime('%Y-%m-%d')

    return formatted_birth_date

clients_exo["birth_year"] = clients_exo.apply(lambda row: calculate_birth_date(row["DrivingLicenseDateObtained"], row["DrivingLicenseAgeObtained"]), axis=1)

clients_exo.head()
```
